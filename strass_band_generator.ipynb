```python
### FUNCTIONS TO CREATE THE ARRAYS THAT CONTAIN THE DATA AND COORDINATES

# adjust the center part to be symmetric if desired
def adjust_symmetry(x, y, b, left, right):
    width = x.shape[1]
    for yy in range(5):
        b[yy, int(left)] = 2
        b[yy, int(width - right)-1] = 2
        for xx in range(int(left), int(width)-int(right)):
            x[yy, xx] += .5
        for xx in range(int(width)-int(right), int(width)):
            x[yy, xx] += 1.
    if x.shape[1] > b.shape[1]:
        b = b[0:b.shape[0]-1, 0:b.shape[1]-2]
    return x, y, b

# populate an array with values that represent the band (stones, voids)
def make_array(initials, width=79., pad=1, symmetric=False):
    adjust = False
    center_part = width_dict[initials[0]] + width_dict[initials[1]] + 2*pad + 1
    print("Center part width will be {:d} for initials {:s}. {:s}.".format(center_part, initials[0], initials[1]))
    border_total = width - center_part
    border_left = int(np.ceil(border_total/2.)) 
    border_right = int(np.floor(border_total/2.))
    if symmetric and border_right != border_left:
        print("Adjusting width for symmetry, reducing it by 1")
        border_left = border_right
        width -= 1
        adjust = True
    print("Left has {:d} columns, center has {:d} columns, right has {:d} columns.".format(
        border_left, center_part, border_right))


    #print(border_left * [letters['I']])
    b = np.concatenate(border_left * [letters['I']], axis=1)
    for ii in range(pad):
        b = np.concatenate([b, letters[' ']], axis=1)
    b = np.concatenate([b, letters[initials[0]], letters[' '], letters[initials[1]]], axis=1)
    for ii in range(pad):
        b = np.concatenate([b, letters[' ']], axis=1)
    br = np.concatenate(border_right*[letters['I']], axis=1)
    b = np.concatenate([b, br], axis=1)
    x, y = np.meshgrid(np.arange(0, width, 1.), np.arange(0, 5., 1.))
    y = np.flip(y, axis=0)
    if adjust:
        return adjust_symmetry(x, y, b, border_left, border_right)
    else:
        return x, y, b
```

```python
# iamge processing library opencv-python
import cv2
import cv2
import urllib.request
import numpy as np

urllist = [
    "https://raw.githubusercontent.com/baeldin/strass_preview_generator/main/strass_void_even.png",
    "https://raw.githubusercontent.com/baeldin/strass_preview_generator/main/strass_single_even.png",
    "https://raw.githubusercontent.com/baeldin/strass_preview_generator/main/strass_void_even_half.png"
]

img_list = [] 

for url_string in urllist:
    req = urllib.request.urlopen(url_string)
    arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
    img_list.append(cv2.imdecode(arr, -1)) # 'Load it as it is'


# merge half and full void to form void of 1.5 times normal width
strass_void_wider = cv2.hconcat([img_list[0], img_list[2]])

# make a list for later reference
# contains void - stone - wider void
strass_img_list = [img_list[0], img_list[1], strass_void_wider]
# show stone, just for testing purposes
plt.imshow(strass_img_list[1])
```

```python
def draw_band(x, y, b, outname="out_test.png"):
    size_y, size_x = x.shape
    list_full = []
    img_list = []
    for yy in range(size_y):
        list_x = []
        for xx in range(size_x):
            list_x.append(strass_img_list[b[yy,xx]])
        img_list.append(cv2.hconcat(list_x))
    img = cv2.vconcat(img_list)
    #fig, ax = plt.subplots(1,1, figsize=(9, .75))
    #ax.scatter(x, y, s=1, c=b, vmin=0, vmax=2)
    #plt.show()

    plt.imshow(img)
    cv2.imwrite(outname, img)
    return img
```

```python
### MAKE YOU BAND HERE:
#
# make_array takes several arguments:
# ['A', 'B'] .... list of two letters, the desired initials
# pad = N ....... integer, minimum colums around initials
# width ......... integer, number of vertical columns total
# symmetric ..... bool (True or False), adjust by one half unit
#                 if left and right borders are not equal
x, y, b = make_array(['R', 'D'], pad=1, width=79, symmetric=False)
img = draw_band(x, y, b, outname="output_long.png")
# output file is written to temporary storage, click directory icon on
# the left to open files, output.png (or whatever name was chosen) is
# stored there, not in a subdirecotry.
# This will be deleted when page is closed, download to keep!
```
