```python
### FUNCTIONS TO CREATE THE ARRAYS THAT CONTAIN THE DATA AND COORDINATES

# adjust the center part to be symmetric if desired
def adjust_symmetry(x, y, b, left, right):
    width = x.shape[1]
    for yy in range(5):
        b[yy, int(left)] = 2
        b[yy, int(width - right)-1] = 2
        for xx in range(int(left), int(width)-int(right)):
            x[yy, xx] += .5
        for xx in range(int(width)-int(right), int(width)):
            x[yy, xx] += 1.
    if x.shape[1] > b.shape[1]:
        b = b[0:b.shape[0]-1, 0:b.shape[1]-2]
    return x, y, b

# populate an array with values that represent the band (stones, voids)
def make_array(initials, width=79., pad=1, symmetric=False):
    adjust = False
    center_part = width_dict[initials[0]] + width_dict[initials[1]] + 2*pad + 1
    print("Center part width will be {:d} for initials {:s}. {:s}.".format(center_part, initials[0], initials[1]))
    border_total = width - center_part
    border_left = int(np.ceil(border_total/2.)) 
    border_right = int(np.floor(border_total/2.))
    if symmetric and border_right != border_left:
        print("Adjusting width for symmetry, reducing it by 1")
        border_left = border_right
        width -= 1
        adjust = True
    print("Left has {:d} columns, center has {:d} columns, right has {:d} columns.".format(
        border_left, center_part, border_right))


    #print(border_left * [letters['I']])
    b = np.concatenate(border_left * [letters['I']], axis=1)
    for ii in range(pad):
        b = np.concatenate([b, letters[' ']], axis=1)
    b = np.concatenate([b, letters[initials[0]], letters[' '], letters[initials[1]]], axis=1)
    for ii in range(pad):
        b = np.concatenate([b, letters[' ']], axis=1)
    br = np.concatenate(border_right*[letters['I']], axis=1)
    b = np.concatenate([b, br], axis=1)
    x, y = np.meshgrid(np.arange(0, width, 1.), np.arange(0, 5., 1.))
    y = np.flip(y, axis=0)
    if adjust:
        return adjust_symmetry(x, y, b, border_left, border_right)
    else:
        return x, y, b
```
